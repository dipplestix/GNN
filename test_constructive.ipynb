```python
from algorithms.greedy_color import greedy_color
from algorithms.dsatur import dsatur
from utilities.data_functions import col_to_edge_list
from utilities.vis_helpers import plot_coloring
from utilities.eval import count_collisions
```

```python
# Load and plot the graph

graph = 'queen5_5.col'
n, edge_list, col_list = col_to_edge_list(f'data/input/COLOR/instances/{graph}')
plot_coloring(edge_list)
```

```python
best = 100
for _ in range(10):
    coloring = greedy_color(edge_list)
    if max(coloring) + 1 < best:
        best = max(coloring) + 1
        best_coloring = coloring
print(f'This produced a coloring with {count_collisions(best_coloring, edge_list)} coliisions and {best} colors')
```

```python
plot_coloring(edge_list, best_coloring)
```
